```python
# ============================================
# Task-02: Pixel Manipulation for Image Encryption
# (Fixed Overflow Issue)
# ============================================

import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

# Step 1: Upload Image
uploaded = files.upload()

for filename in uploaded.keys():
    img_path = filename

# Read the image
image = cv2.imread(img_path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Show original image
plt.imshow(image)
plt.title("Original Image")
plt.axis("off")
plt.show()

# Step 2: Encryption Function
def encrypt_image(img, key):
    encrypted = (img.astype(np.int32) + key) % 256
    return encrypted.astype(np.uint8)

# Step 3: Decryption Function
def decrypt_image(img, key):
    decrypted = (img.astype(np.int32) - key) % 256
    return decrypted.astype(np.uint8)

# Key for encryption
secret_key = 50  # you can change this

# Encrypt the image
encrypted_img = encrypt_image(image, secret_key)

# Decrypt the image
decrypted_img = decrypt_image(encrypted_img, secret_key)

# Step 4: Show Results
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
axes[0].imshow(image)
axes[0].set_title("Original")
axes[0].axis("off")

axes[1].imshow(encrypted_img)
axes[1].set_title("Encrypted")
axes[1].axis("off")

axes[2].imshow(decrypted_img)
axes[2].set_title("Decrypted")
axes[2].axis("off")

plt.show()

```
